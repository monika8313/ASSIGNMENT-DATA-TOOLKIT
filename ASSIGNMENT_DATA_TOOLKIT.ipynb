**2. How does broadcasting work in NumPy?**

- Broadcasting is a mechanism in NumPy that allows you to perform arithmetic operations on arrays of different shapes.
- When performing an operation between two arrays, NumPy compares their shapes element-wise, starting from the trailing dimensions.
- Two dimensions are compatible for broadcasting if they are equal, or if one of them is 1.
- If the dimensions are incompatible, an error is raised.
- If one array has fewer dimensions than the other, its shape is padded with ones on the left.
- If a dimension of one array is 1 and the corresponding dimension of the other array is greater than 1, the dimension with size 1 is stretched or "broadcast" to match the size of the other dimension.

**Example:**

```python
import numpy as np

# Example 1: Scalar and array
a = np.array([1, 2, 3])
b = 2
result1 = a + b  # Broadcasting the scalar 'b' to the shape of 'a'
print("Example 1 (Scalar and array):", result1)

# Example 2: Arrays of different shapes
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([10, 20, 30])
result2 = c + d  # Broadcasting 'd' to match the rows of 'c'
print("Example 2 (Arrays of different shapes):", result2)

# Example 3: Arrays with a dimension of size 1
e = np.array([[1], [2], [3]])
f = np.array([10, 20, 30])
result3 = e + f # Broadcasting 'e' and 'f'
print("Example 3 (Arrays with a dimension of size 1):", result3)
```

**6. What are the differences between NumPy arrays and Python lists?**

- Python Lists: Can hold mixed data types (e.g., integers, strings, objects).
- NumPy Arrays: Require elements to be of the same data type, which improves performance and memory efficiency.
2. Performance
- Python Lists: Slower for numerical computations because operations must be performed element-wise via loops or comprehensions.
- NumPy Arrays: Faster due to optimized, precompiled C code underneath and support for vectorized operations.
3. Memory Usage
- Python Lists: Consume more memory because they store references to objects.
- NumPy Arrays: More memory-efficient since they store data in contiguous blocks.
4. Functionality
- Python Lists: Limited built-in capabilities for numerical operations.
- NumPy Arrays: Rich suite of mathematical functions—element-wise operations, linear algebra, statistical tools, and more.
5. Indexing and Slicing
- Both support slicing, but:
- NumPy Arrays: Offer more advanced indexing (e.g., boolean masking, multi-dimensional slicing).
- Lists: Have basic slicing and indexing.
6. Multidimensional Support
- Python Lists: Can simulate multidimensional arrays using nested lists, but handling them gets complex.
- NumPy Arrays: Designed for multidimensional arrays; operations like matrix multiplication are easy and efficient.



Differences between Python lists and NumPy arrays:

- Python lists can hold mixed data types, while NumPy arrays require homogeneous data types.
- Python lists are slower for numerical operations; NumPy arrays are much faster due to vectorized computations.
- Python lists use more memory, storing data as individual object references. NumPy arrays are more memory-efficient with contiguous storage.
- Python lists offer basic functionality for math; NumPy arrays have extensive built-in mathematical and statistical operations.
- Both support slicing, but NumPy arrays allow for advanced indexing, including boolean masks and multidimensional slicing.
- Python lists simulate multidimensional data through nesting, which can get messy. NumPy arrays natively support multiple dimensions with clean syntax and powerful operations.

**8. What does the term “vectorized operation” mean in NumPy?**

 -In NumPy, a vectorized operation refers to performing operations on entire arrays (or large chunks of data) without using explicit Python loops. Instead of processing elements one at a time, NumPy uses highly optimized, low-level C code to apply operations across entire arrays simultaneously. This leads to significantly faster and more efficient computation.


Key Benefits:

- Speed: Much faster than using for loops, especially on large datasets.
- Concise Code: Fewer lines of code make logic easier to read and maintain.
- Efficiency: Reduces overhead from Python’s interpreted loop execution.


Example without Vectorization (Using a loop):


```python
import numpy as np

a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
result = []

for i in range(len(a)):
    result.append(a[i] + b[i])

print(result)
```

```python
#Example with Vectorized Operation

import numpy as np

a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
result = a + b

print(result)
```

NumPy supports vectorized operations for addition, subtraction, multiplication, division, exponentiation, and many mathematical functions (np.sin, np.exp, np.mean, etc.).

**13. Why is handling missing data important in Pandas?**

 -Handling missing data in Pandas is crucial because unaddressed gaps can distort your analysis, lead to incorrect conclusions, or break your code during computations. Here’s why it matters:

- Preserves Data Integrity: Missing values can skew summary statistics or aggregations. Managing them ensures your results remain reliable.

- Prevents Errors in Analysis: Many Pandas operations (like mean, sum, or plotting) may fail or behave unexpectedly with NaNs present.

- Improves Model Performance: In machine learning workflows, missing data can affect feature selection and predictive accuracy if not properly imputed or removed.

- Enables Clean Visualizations: Charts and plots can become misleading or incomplete if missing values aren't handled.

- Supports Effective Decision-Making: Clean, complete datasets lead to better insights, especially in critical fields like healthcare, finance, or policy.


Typical strategies include:

- Removing missing entries using dropna()
- Filling gaps with statistical values like mean or median using fillna()
- Interpolation for time-series data
- Flagging rows or columns with missing data to inform decisions


```python
import pandas as pd
import numpy as np

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [25, np.nan, 30, 22, 28],
    'Score': [85, 90, np.nan, 88, 95]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Drop rows with any missing values
df_dropped = df.dropna()
print("\nDataFrame after dropping rows with missing data:")
print(df_dropped)

# Fill missing values
df_filled = df.fillna({
    'Name': 'Unknown',
    'Age': df['Age'].mean(),
    'Score': df['Score'].median()
})
print("\nDataFrame after filling missing data:")
print(df_filled)
```

This covers:
- Identifying missing values with isnull()
- Removing incomplete rows with dropna()
- Imputing missing values using mean, median, or placeholder strings with fillna()

**17.  Explain the difference between apply() and map() in Pandas?**

 -In Pandas, apply() and map() are both used to apply functions to data, but they serve slightly different purposes and work on different data structures.

map()

- Used only on Series objects (not DataFrames).
- Applies a function element-wise to each value in the Series.
- Ideal for simple transformations, like scaling or renaming.

- Can accept:

 - A function (e.g., lambda x: x + 1)
 - A dictionary (useful for mapping values)
 - A Series (for alignment-based substitution)


Example:


```python
import pandas as pd

s = pd.Series([1, 2, 3])
s_mapped = s.map(lambda x: x ** 2)
print(s_mapped)
```

apply()

- Works on both Series and DataFrames.
- For Series, similar to map()—applies a function element-wise.
- For DataFrames, the function is applied to either rows (axis=1) or columns (axis=0).
- Useful for complex operations involving multiple columns or aggregation.


Example on Series:



```python
s_applied = s.apply(lambda x: x ** 2)
```

```python
df = pd.DataFrame({'a': [1, 2], 'b': [3, 4]})
df_applied = df.apply(lambda x: x.sum(), axis=0)
```

Summary

- Use map() for straightforward element-wise transformations on a Series.
- Use apply() for flexible and complex operations—especially on DataFrames or when dealing with multiple columns.

**21. Why is NumPy’s array slicing faster than Python’s list slicing?**

 -NumPy’s array slicing is faster than Python’s built-in list slicing primarily because of how NumPy stores and processes data under the hood:


- Contiguous Memory Layout: NumPy arrays are stored in a contiguous block of memory, which enables highly efficient access and manipulation. Python lists, on the other hand, are arrays of pointers to objects scattered across memory, introducing overhead.

- Homogeneous Data Types: NumPy arrays require all elements to be of the same data type. This allows for predictable and compact memory usage, while Python lists can hold mixed types, which adds complexity and slows down operations.

- Built-in Optimizations: NumPy is implemented in C under the hood. Its slicing operations are backed by optimized compiled code, making them much faster than Python’s interpreted list operations.

- Views Instead of Copies: When you slice a NumPy array, it returns a view on the original data, not a full copy. Python lists return a completely new list when sliced, which takes time and memory.

- Vectorized Access Patterns: NumPy arrays support advanced slicing with steps, multi-dimensional access, and broadcasting—all performed in optimized low-level routines without explicit loops.


This speed difference becomes especially noticeable when working with large datasets or running numerical simulations. If you're dealing with high-performance data manipulation, NumPy is the clear winner.
